#### Crear External Locations requeridas para el Proyecto
1. bronze
2. silver
3. gold

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS sadatabricksext_bronze
    URL 'abfss://bronze@dbsadatabricksext.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL `databricks-ext-storage-credendial`);
```

```sql
DESC EXTERNAL LOCATION sadatabricksext_bronze;
```

Valido acceso

```sql
%fs
ls "abfss://bronze@dbsadatabricksext.dfs.core.windows.net/"
```
